```python
import pandas as pd
dfN = pd.read_csv('Trump_nodes.csv')
dfE = pd.read_csv('Trump_edges.csv')
```

```python
dfN
```

```python
dfE
```

```python
dfN['SeqId']=range(1,len(dfN)+1)
```

```python
dfNsimple=dfN.drop(['Label','Followers','Lang'],axis=1)
```

```python
dfNsimple
```

```python
dfT=pd.merge(dfE, dfNsimple, left_on='Source', right_on='Id')
```

```python
dfT
```

```python
dfT['Source']= dfT['SeqId']
del dfT['SeqId']
del dfT['Id']
```

```python
dfT
```

```python
dfEout=pd.merge(dfT, dfNsimple, left_on='Target', right_on='Id')
```

```python
dfEout
```

```python
dfEout['Target']= dfEout['SeqId']
del dfEout['SeqId']
del dfEout['Id']
```

```python
dfEout
```

```python
dfEout.to_csv('edgesSeq.csv', sep=',', index = False)
```
